## Taller `Matrices`

```python
import sympy as sy

sy.init_printing()

a, b = sy.symbols("alpha, beta")

A = sy.Matrix([[a, b],[b, -a]])

A
```

```python
A.eigenvals()
```

```python
A.eigenvects()
```
