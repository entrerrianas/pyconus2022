# Dibujar un diagrama

### El procedimiento mínimo para dibujar un diagrama de Venn es el siguiente.

> https://recursospython.com/guias-y-manuales/diagramas-de-venn-con-matplotlib/

```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn2
```

```python
venn2((4, 2, 1))
```

#### Argumentos: (a, b, c) : a = tamaño Diagrama A. b = tamaño Diagrama B. c = tamaño Interseccion A y B.
#### En el ejemplo a = 4, b = 2, c = 1. a == Diagrama A. b== Diagrama B. c == Interseccion A y B.

> El argumento indica el tamaño relativo de los tres subconjuntos en este orden: 
**Ab (izquierda)**, **aB (derecha)**, **AB (intersección).** 
Así, la tupla (4, 2, 1) dibujaría el conjunto A el doble de tamaño respecto de B.

> Para identificar a cada uno de los subconjuntos (tres en diagramas de dos conjuntos) el módulo utiliza una nomenclatura que consiste en colocar un 1 para indicar que la sección está incluida en el conjunto y un 0 para indicar que está excluida. De esta manera, siguiendo el orden **«ABC»**, el subconjunto 10 es el de la izquierda (el que pertenece a A pero no a B); el 01, el de la derecha (el que pertenece a B pero no a A); y el 11, el del medio (la intersección).

```python
# set_colors determina el color de los conjuntos.

venn2((1, 1, 1), set_colors=("#888888", "#0000FF"))

# El color de las instersecciones es calculado automáticamente.
```

```python
# Nótese que, por defecto, tienen un valor alpha de 0.4, i.e., 60% de transparencia.
# Utilícense valores de 0.0 a 1.0 en el parámetro de ese mismo nombre.
# Deshabilita la transparencia.

venn2((1, 1, 1), set_colors=("#888888", "#0000FF"), alpha=1)
```

```python
# La función venn2() retorna una instancia de una clase llamada VennDiagram.
# En particular, los métodos de ésta que nos interesan son get_label_by_id() y get_patch_by_id().
# Ambas toman un subconjunto y retornan instancias de matplotlib.text.Text y matplotlib.patches.PathPatch, respectivamente.
# A partir de ellas podemos modificar individualmente el aspecto de cada subconjunto.
# Por ejemplo, el siguiente código cambia el color y el texto del subconjunto 11 (i.e. AB).

diagrama_venn = venn2((4, 2, 1))
diagrama_venn.get_label_by_id("11").set_text("Intersección")
diagrama_venn.get_label_by_id("11").set_color("#FFFFFF")
```

```python
# O bien alterar el color de un subconjunto en particular: PARA CALCULAR LA MEZCLA de COLORES,
# como la formacion apartir de los colores primarios


diagrama_venn = venn2((4, 2, 1))
diagrama_venn.get_label_by_id("11").set_text("Intersección")
diagrama_venn.get_label_by_id("11").set_color("#FFFFFF")

# Establece el color de cada Diagrama A y B.
diagrama_venn.get_patch_by_id("10").set_color('y') # yellow : amarillo
diagrama_venn.get_patch_by_id("01").set_color('b') # blue : azul

# Establece el color de la intersección:
diagrama_venn.get_patch_by_id("11").set_color('g') # green : verde (inteserccion entre amarillo y azul)
```

```python
# El parámetro set_labels permite cambiar los nombres de los conjuntos mostrados en el diagrama.

venn2((1, 1, 1), set_labels=("ROJO", "AZUL"))
```
